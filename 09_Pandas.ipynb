```python
df_pokedex['total_attack'] = df_pokedex['attack'] + df_pokedex['sp_attack']
df_pokedex.head()
```

```python
df_pokedex.get(['hp', 'attack', 'defense',	'sp_attack', 'sp_defense', 'speed']).sum(axis = 1)
```

```python
df_pokedex['total_stats'] = df_pokedex.get(['hp', 'attack', 'defense',	'sp_attack', 'sp_defense', 'speed']).sum(axis = 1)
df_pokedex.head()
```

```python
# apply
df_pokedex['generation'].unique()
```

```python
def generaciones(numero):
    if numero == 1:
        return 'Kanto'
    elif numero == 2:
        return 'Johto'
    elif numero == 3:
        return 'Hoenn'
    elif numero == 4:
        return 'Sinnoh'
    elif numero == 5:
        return 'Unova'
    elif numero == 6:
        return 'Kalos'
    elif numero == 7:
        return 'Alola'
df_pokedex['region'] = df_pokedex['generation'].apply(generaciones)
df_pokedex.head()
```
